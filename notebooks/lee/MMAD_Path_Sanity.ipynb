# MMAD Path Sanity Check

`mmad.json`과 몇 장의 이미지/마스크가 정상적으로 열리는지 빠르게 확인합니다.

```python
import os, json, random
from pathlib import Path
from PIL import Image

MMAD_ROOT = Path(os.environ.get("MMAD_ROOT", "/Volumes/T7/Dataset/MMAD")).expanduser()
MMAD_JSON = MMAD_ROOT / "mmad.json"

mmad = json.loads(MMAD_JSON.read_text(encoding="utf-8"))
print("num images:", len(mmad))

# 랜덤 5개 점검
for k in random.sample(list(mmad.keys()), 5):
    meta = mmad[k]
    img_rel = meta.get("image_path") or k
    mask_rel = meta.get("mask_path")
    img_p = MMAD_ROOT / img_rel
    ok_img = img_p.exists()
    ok_mask = bool(mask_rel) and (MMAD_ROOT / mask_rel).exists()
    size = None
    if ok_img:
        with Image.open(img_p) as im:
            size = im.size
    print(f"- {img_rel} | img={ok_img} size={size} | mask={ok_mask} ({mask_rel})")
```
