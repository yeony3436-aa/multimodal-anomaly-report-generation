### MMAD Json

```python
def parse_source_and_category(rel_path: str) -> tuple[str, str]:
    # 예: DS-MVTec/bottle/image/...  -> source=DS-MVTec, category=bottle
    parts = rel_path.split("/")
    source = parts[0] if len(parts) > 0 else "unknown"
    category = parts[1] if len(parts) > 1 else "unknown"
    return source, category

def normalize_options(opts) -> dict:
    # opts가 dict(A->text)거나 list 등일 수 있음. dict 형태로 정규화.
    if isinstance(opts, dict):
        return {str(k).strip(): str(v).strip() for k,v in opts.items()}
    if isinstance(opts, list):
        out = {}
        for i, v in enumerate(opts):
            out[chr(ord("A")+i)] = str(v).strip()
        return out
    return {"A": str(opts)}

def get_image_size(rel_path: str) -> tuple[int, int]:
    """이미지 크기 (width, height) 반환. 파일 없으면 (None, None)"""
    try:
        img_path = DATA_ROOT / rel_path
        with Image.open(img_path) as img:
            return img.size  # (width, height)
    except Exception:
        return (None, None)

rows = []
for k, meta in tqdm(mmad_json.items(), desc="Loading images"):
    rel_img = k
    src, cat = parse_source_and_category(rel_img)
    mask = meta.get("mask_path")
    sim_t = meta.get("similar_templates", []) or []
    rnd_t = meta.get("random_templates", []) or []
    width, height = get_image_size(rel_img)

    for q in meta.get("conversation", []):
        question = q.get("Question") or q.get("question") or ""
        answer   = q.get("Answer") or q.get("answer") or ""
        qtype    = q.get("type") or q.get("question_type") or "unknown"
        opts     = normalize_options(q.get("Options") or q.get("options") or {})
        rows.append({
            "query_image": rel_img,
            "source": src,
            "category": cat,
            "mask_path": mask,
            "num_similar_templates": len(sim_t),
            "num_random_templates": len(rnd_t),
            "question_type": qtype,
            "question": question,
            "answer": str(answer).strip(),
            "num_options": len(opts),
            "options_keys": "".join(sorted(opts.keys())),
            "img_width": width,
            "img_height": height,
        })

df = pd.DataFrame(rows)
print(df.shape[0])
df.head()
```

```python
import random
from re import A
from PIL import Image
import matplotlib.pyplot as plt


def show_sample(rel_img: str, template_kind: str = "similar"):
    """
    이미지 샘플 시각화 (Query, Mask, Template)
    - rel_img: mmad_json의 key (상대 경로)
    - template_kind: 'similar' 또는 'random'
    """
    # mmad_json에서 메타데이터 조회
    meta = mmad_json.get(rel_img)
    if meta is None:
        print("meta not found for:", rel_img)
        return

    # Query 이미지 경로
    query_path = DATA_ROOT / rel_img

    # Mask 경로 (없을 수 있음 - 정상 이미지)
    mask_path = None
    if meta.get("mask_path"):
        # rel_img에서 상위 2개 디렉토리 추출 (예: image/broken_large)
        base_dir = Path('/'.join(rel_img.split('/')[:2]))
        mask_path = DATA_ROOT / base_dir / meta.get("mask_path")

    # Template 경로
    tmpl_list = meta.get("similar_templates") if template_kind == "similar" else meta.get("random_templates")
    tmpl_path = (DATA_ROOT / tmpl_list[0]) if tmpl_list else None

    # 시각화
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    # Query 이미지
    axes[0].set_title("Query")
    if query_path.exists():
        axes[0].imshow(Image.open(query_path))
    else:
        axes[0].text(0.5, 0.5, "Not found", ha="center", va="center")
    axes[0].axis("off")

    # Mask 이미지
    axes[1].set_title("Mask")
    if mask_path and mask_path.exists():
        axes[1].imshow(Image.open(mask_path))
    else:
        axes[1].text(0.5, 0.5, "No mask", ha="center", va="center")
    axes[1].axis("off")

    # Template 이미지
    axes[2].set_title(f"Template ({template_kind})")
    if tmpl_path and tmpl_path.exists():
        axes[2].imshow(Image.open(tmpl_path))
    else:
        axes[2].text(0.5, 0.5, "No template", ha="center", va="center")
    axes[2].axis("off")

    plt.suptitle(rel_img, fontsize=10)
    plt.tight_layout()
    plt.show()


# 무작위 샘플 2개 시각화
keys = list(mmad_json.keys())
for rel in random.sample(keys, 3):
    show_sample(rel, template_kind="similar")
```
